```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 실제 데이터가 담겨있는 소스가 다른 곳에 있을 경우가 많음. 
#이럴 때 iframe으로 찾기
res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=201641&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=sympathyScore&page=2")
```

```python
soup = bs(res.text, "lxml")
```

```python
# 영화 관람객들의 리뷰데이터 수집
review = soup.select("div.score_reple > p")
```

```python
for i in review:
    print(i.text.strip())
```

```python
viewer = soup.select("span.ico_viewer")
```

```python
# 불필요한 데이터 추출해서 제거하는 방법
# extract() - > 원본데이터에 영향을 준다
for i in viewer:
    i.extract()
```

```python
for i in review:
    print(i.text.strip())
```

```python
spo = soup.select("span.text_spo._text_spo")
```

```python
for i in spo:
    i.extract()
```

```python
for i in review:
    print(i.text.strip())
```

```python
for i in tq(range(1,50)):
    res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=201641&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=sympathyScore&page="+str(i))
    soup = bs(res.text, "lxml")
    viewer = soup.select("span.ico_viewer")
    spo = soup.select("span.text_spo._text_spo")
    for j in viewer:
        j.extract()
    for p in spo:
        p.extract()
    review = soup.select("div.score_reple > p") 
    for t in review:
        print(t.text.strip())
```
