```python
# comp_ass1 : 직원들이 일을 잘 했을 때 인정을 받고 칭찬을 듣는다
# comp_ass2 : 경영진은 직원들이 업무를 잘 수행한다고 믿는다
# comp_ass3 : 갈등은 공정한 방식으로 처리된다
# comp_ass4 : 업무가 공평하게 분배된다
# comp_ass5 : 나와 동료 사이에 협력이 잘 된다
# comp_ass6 : 일반적으로 직원들은 경영인을 신뢰한다
# 1 매우 동의한다 2 대체로 동의한다 3 보통이다 
# 4 대체로 동의하지 않는다 5 전혀 동의하지 않는다

variable['comp_ass1'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['comp_ass2'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['comp_ass3'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['comp_ass4'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['comp_ass5'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['comp_ass6'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
```

```python
# wsituation11 : 직장에서 공정하게 대우를 받는다
# wsituation1 : 동료들은 나를 도와주고 지지해준다
# wsituation2 : 상사는 나를 도와주고 지지해준다
# 1 항상 그렇다 2 대부분 그렇다 3 가끔 그렇다 
# 4 별로 그렇지 않다 5 전혀 그렇지 않다
# 등차간격으로 하면 안 될 것 같음!

variable['wsituation11'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['wsituation1'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['wsituation2'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
```

```python
# manaqual1 : 직속상관은 나를 인격적으로 존중해준다
# manaqual2 : 직속상관은 직원들이 함께 일을 잘할 수 있도록 돕는다
# manaqual3 : 일ㅇ르 처리하는 데 도움이 된다
# manaqual4 : 직속상관은 일에 대해 도움이 되는 조언(피드백)을 해준다
# manaqual5 : 직속상관은 내가 발전하도록 격려하고 도와준다
# 1 매우 동의한다 2 대체로 동의한다 3 보통이다 
# 4 대체로 동의하지 않는다 5 전혀 동의하지 않는다

variable['manaqual1'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['manaqual2'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['manaqual3'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['manaqual4'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['manaqual5'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
```

```python
# wsituation3 : 작업 목표가 결정되기 전에 나의 의견을 묻는다
# wsituation4 : 부서, 조직의 구성이나 업무 절차 개선에 참여한다
# wsituation5 : 같이 일할 사람(동료)을 선택할 때 나의 의견이 반영된다
# wsituation9 : 일을 할 때 내 생각을 반영할 수 있다
# wsituation13 : 내가 하는 일의 중요한 결정에 영향을 미칠 수 있다
# 1 항상 그렇다 2 대부분 그렇다 3 가끔 그렇다 
# 4 별로 그렇지 않다 5 전혀 그렇지 않다
# 등차간격으로 하면 안 될 것 같음!

variable['wsituation3'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['wsituation4'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['wsituation5'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['wsituation9'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
variable['wsituation13'].replace([1, 2, 4, 5], [5, 4, 2, 1], inplace=True)
```
