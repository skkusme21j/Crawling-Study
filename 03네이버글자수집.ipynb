```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res =req.get("https://www.naver.com/")
```

```python
res.text
```

```python
soup = bs(res.text, "lxml")
```

```python
#soup이라는 데이터에서 메일, 카페, 블로그 같은 글자들만 수집해줘
#soup.select("선택자")
naver = soup.select("a.nav")
```

```python
naver[0].text
```
