```python
# 영상의 제목을 수집
title = driver.find_elements(By.ID, "video-title")
```

```python
driver = wb.Chrome()
time.sleep(0.5)
driver.get("https://www.youtube.com/channel/UCJpAwvQaZyCI5spAz7tipGA/videos")
time.sleep(1)
body = driver.find_element(By.CSS_SELECTOR,'body')
for i in range(10):
    body.send_keys(Keys.END)
    time.sleep(1)
     
title = driver.find_elements(By.ID, "video-title")
view = driver.find_elements(By.CSS_SELECTOR,"div#metadata-line > span:nth-child(1)") 
titleList = []
viewList = []
for i in range(len(title)):
    titleList.append(title[i].text)
    viewList.append(view[i].text)
len(titleList)
len(viewList)
print(titleList)
driver.quit()
```
